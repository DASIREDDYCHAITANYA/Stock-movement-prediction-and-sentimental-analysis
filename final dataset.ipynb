```python
dataset1='/content/drive/MyDrive/dhoni/dataset1.csv'
dataset2='/content/drive/MyDrive/dhoni/merged_data.csv'

```

```python
import pandas as pd
dataset1 = pd.read_csv(dataset1)
dataset2 = pd.read_csv(dataset2)







dataset1['time_published'] = pd.to_datetime(dataset1['time_published'])

dataset1 = dataset1.groupby('time_published').agg({'overall_sentiment_score': 'mean'}).reset_index()
```

```python
dataset1
```

```python
dataset1 = dataset1.rename(columns={'time_published	': 'date', 'overall_sentiment_score': 'sentiment_score'})
dataset2 = dataset2.rename(columns={'date': 'time_published'})
```

```python
dataset2
```

```python
dataframes = [dataset1, dataset2]


merged_df = pd.concat(dataframes, axis=0, ignore_index=True)


merged_df
```

```python
df3="/content/drive/MyDrive/dhoni/tesla.csv"
df3=pd.read_csv(df3)
df3
```

```python

merged_df['time_published'] = pd.to_datetime(merged_df['time_published'])
df3['Date'] = pd.to_datetime(df3['Date'])

final_df = pd.merge(merged_df, df3, left_on='time_published',right_on="Date")
```

```python
final_df
```

```python
# Define the path where you want to save the file
file_path = '/content/drive/MyDrive/dhoni/final_dataset.csv'

# Save the DataFrame to a CSV file
final_df.to_csv(file_path, index=False)
```
